```python
import pygrib
import matplotlib.pyplot as plt
```

### download a '.grib2' file

https://console.cloud.google.com/storage/browser/high-resolution-rapid-refresh;tab=objects?project=gcp-public-data-weather&prefix=&forceOnObjectsSortingFiltering=true


```python
#load a '.grib2' file
grbs = pygrib.open('hrrr.20151130_conus_hrrr.t00z.wrfnatf13.grib2')
```

```python
# have a look on what it contains.
for grb in grbs():
    print(grb)
```

```python
# print the category related to certain keywords
def find(*kws):
    for kw in kws:
        print(kw)
        for grb_name in grbs():
            if kw.lower() in str(grb_name).lower():
                print(grb_name)
find("snow")
```

```python
# plot this category
grb = grbs()[9-1]
mat = grb.values
imgplot = plt.imshow(mat)
plt.colorbar()
plt.title(grb.name)
```
