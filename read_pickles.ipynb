```python
import os
import datetime
import pygrib
import pickle
import numpy as np
import pygrib
import matplotlib.pyplot as plt


def series_show(X):
    for i in range(12):
        fig, axs = plt.subplots(1,8,figsize=(15,15))
        
        for j in range(8):
            time = (i*2 + j//4)*100 + j%4*15
            axs[j].imshow(X[time])
            axs[j].set_title('{:<4}'.format(time))
        plt.show()


cur_location = os.getcwd()

with open(cur_location+'/hrrrdata/hrrr2021-04-15.dat','rb') as f:
    (cloudtop, VIL, names) = pickle.load(f)
```

```python
series_show(cloudtop)
```

```python
series_show(VIL)
```
